## Extracting Twitter information

```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/drive/MyDrive/data4origscrap.csv',encoding='latin-1')
df=df.drop(['Unnamed: 0', 'Unnamed: 0.1','Unnamed: 0.1.1'],axis=1)
df.head(10)
```

```python
ahat=[];bhat=[]
for x in range(df.shape[0]):
  ahat.append(str(df.iat[x,3]).replace(' ','+')+'+'+str(df.iat[x,2]).replace(' ','+'))
  bhat.append(str(df.iat[x,3]).replace(' ','+')+'+'+str(df.iat[x,2]).replace(' ','+'))
print(len(ahat))
print(len(bhat))
```

```python
for i in range(len(ahat)):
  if(i<5):
    print(bhat[i])
  
```

```python
import requests
import bs4

t=[]
aut=[]
for i in range(12600,13300):
  
  print(len(t),' ',i)

  a=str(ahat[i])
  b=str(bhat[i])
  text= a + "+twitter"
  url = 'https://google.com/search?q=' + text
  

  try:
    request_result=requests.get(url)

    # Creating soup from the fetched request
    soup = bs4.BeautifulSoup(request_result.text,
                "html.parser")

    # soup.find.all( h3 ) to grab
    # all major headings of our search result,
    heading_object=soup.find_all( 'h3' )
  except Exception as e:
    print("conncetion lost")
    continue

  for info in heading_object:
    if(len(info.getText().split('|'))==2):
      if(info.getText().split('|')[1]==' Twitter'):
        t.append(info.getText().split('(')[1].split(')')[0])
        aut.append(b)
```
