######2) Removing scientists having < 5 publications

```python
import pandas as pd

df = pd.read_csv('drive/MyDrive/pubmed_data.csv',dtype=object)
df.head(10)
```

```python
df=df.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1)
df.head(10)
```

```python
dups = df.pivot_table(index = ['name'], aggfunc ='size')
```

```python
df1= dups.to_frame()
df1.head(10)
```

```python
df1['name'] = df1.index
```

```python
df1.columns=['number','name1']
```

```python
df1 = df1.reset_index()
df1.head(10)
```

```python
df1= df1.drop(['name1'],axis=1)
df1.head(10)
```

```python
df2 = df1[df1['number']>=5]
df2.head(10)
```

```python
df2.reset_index
df2.to_csv('pubmed_new.csv')
```
